```python
# manage imports
import pandas as pd
import os
import itertools
import collections
from math import isnan
```

```python
# provide the file name with extension
file_name = "Cog Distance 20190923 v2.xlsx"
# direct to the file path
file_path = os.getcwd() + "/" + file_name
print("Location to file: " + file_path)
```

```python
# now, we open the file directed in the file_path variable
xlsx = pd.ExcelFile(file_path)

# get the first sheet
sheet1 = xlsx.parse(0)
df = sheet1
# get the first column as a list you can loop through
# where the is 0 in the code below change to the row or column number you want    
column = sheet1.iloc[:,0]

# get the first row as a list you can loop through
row = sheet1.iloc[0,:]

#df['Genre4'].value_counts()
# Now, we count the total number of all the different genres in the file
# Then we put it in a dictionary
genre_dictionary = pd.value_counts(df.values.ravel()).to_dict()

# Now, we find out all possible combinations of the genres (taking 2 at a time)
combo_list = list(itertools.combinations(genre_dictionary, 2))
print("Total number of possible combinations:", len(combo_list))

print("Genre:" + combo_list[0][1])

print(genre_dictionary)
```

```python
# find all possible combinations of the genre tuples

combo = collections.Counter([str(k) for l in df.apply(
    lambda x: [c for i in range(2, 3) for c in itertools.combinations(x, i)], axis=1).values for k in l])
# filtered_list = [tup for tup in list1 if len(tup) > 1]
print(combo)
```

```python
counter_dict = dict(combo)
for key, value in counter_dict.items() :
    print(key, ", Value: ", value)
```

```python
#print(counter_dict.get("('Action', 'Adventure')"))
```

```python
cognitive_distance = []
for i in range(len(df)):
    temp = 0
    div = 0
    for j in range(6):    
        if(str(df.iloc[i]['Genre'+str(j+2)]) != "nan"):
            dict_key_1 = df.iloc[i]['Genre1']
            dict_key_2 = df.iloc[i]['Genre'+str(j+2)]
            # print("('" + dict_key_1 + "', '" + dict_key_2 + "')")
            key = "('" + dict_key_1 + "', '" + dict_key_2 + "')"
            cog_val = 1 - (counter_dict.get(key)/genre_dictionary.get(dict_key_1))
            temp += cog_val
            div += 1
        elif(str(df.iloc[i]['Genre2']) == "nan"):
            temp = 0
            div = 1
    cognitive_distance.append(temp/div)
```

```python
df["Cognitive Distance"] = cognitive_distance
print(df)
df.to_excel("updated_" + file_name)
```
